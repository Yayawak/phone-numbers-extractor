```python
def generate_phone_numbers(prefix, remaining_digits, max_length=12):
    """
    Generates a nary-tree of phone numbers in the format 83-xxx-xxxx, recursively.

    Args:
        prefix (str): The current phone number prefix.
        remaining_digits (int): The number of digits left to generate.
        max_length (int): The maximum length of the phone number (default: 12).

    Returns:
        list: A list of generated phone numbers.
    """

    if remaining_digits == 0:
        return [prefix]  # Base case: return complete phone number

    numbers = []
    for digit in range(10):
        new_prefix = prefix + str(digit)
        if len(new_prefix) <= max_length:  # Ensure length limit
            numbers.extend(generate_phone_numbers(new_prefix, remaining_digits - 1, max_length))

    return numbers

# Example usage:
# phone_numbers = generate_phone_numbers("83", 8)  # Generate 8-digit numbers starting with 83
phone_numbers = generate_phone_numbers("83", 2)  # Generate 8-digit numbers starting with 83
print(phone_numbers)
```

```python
def permute(s):
  out = []
  if len(s) == 1:
    return s
  else:
    for i,let in enumerate(s):
      for perm in permute(s[:i] + s[i+1:]):
        out += [let + perm]
  return out

# l = permute(['1','2','3','4'])
# l = [str(x) for x in range(1, 10)]
l = [str(x) for x in range(1, 5)]
# l = permute(['1','2','3','4', '5'])
l = permute(l)
for ll in l:
    print(ll, end=", ")
```
